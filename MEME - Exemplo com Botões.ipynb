```python
#
# Bibliotecas usadas
#

from ipywidgets import interact, widgets, interactive, fixed, interact_manual, FileUpload
from IPython.display import clear_output

import matplotlib.pyplot as plt # gráficos

import numpy as np  # matrizes e matemática

import seaborn as sns #gráficos para análise de dados
plt.ioff() #sem isto os gráficos não "desaparecem" quando se "limpa"
```

```python
#
# uma ou mais funções que implementam o conteúdo científico
#
def regressao_linear(xtxt='1.1 2 3.5', ytxt='4.2 5 6.7', citxt='95'):
    """
    Obtém os pontos do "texto" para uma "python list" (um vetor).
    """

    # Produz dois vetores a partir do texto contendo números
    x_vector = np.array([ float(t) for t in xtxt.split()])
    y_vector = np.array([ float(t) for t in ytxt.split()])

    # Produz um gráfio de regressão usando a biblioteca "seaborn"
    #
    # https://seaborn.pydata.org/tutorial/regression.html
    #

    rp = sns.regplot(x=x_vector, y=y_vector, ci=float(citxt)) 
    
    return rp
```

```python
# botões

x_texto = widgets.Textarea(
    value='1.1 2.1 3.3',
    placeholder='1.1 2.1 3.3',
    description='Valores x:',
    disabled=False
)

y_texto = widgets.Textarea(
    value='4.1 6.3 7.8',
    placeholder='4.1 6.3 7.8',
    description='Valores y:',
    disabled=False
)


ci_values = widgets.Dropdown(
    options=['90', '95', '99'],
    value='95',
    description='Grau de Confiança:',
    disabled=False,
)


#
# Carregar ficheiros xlsx  (ainda não ativo)
#
upload = FileUpload(accept='.xlsx', multiple=True)


#
# Output field
# https://ipywidgets.readthedocs.io/en/latest/examples/Output%20Widget.html
#
output = widgets.Output(layout={'border': '1px solid black'})

#
# Botão Mostra
#
def show_button_handle(btn):
    #output.clear_output()
    #return_value = slider.value
    with output:
        #display(x_texto.value)
        #display(y_texto.value)
        
        clear_output()
        
        #mostrar um gráfico
        display( plt.show(regressao_linear(x_texto.value, y_texto.value, ci_values.value)) );
        
        #mostrar estatísticas
        print("Coeficientes:", "(aqui coeficientes)" )
        
        
   # associa o click no botão Mostra à função show_button_handle
show_button = widgets.Button(description="Mostra")
show_button.on_click(show_button_handle)

#
# Botão Limpa
#
def clear_button_handle(btn):
    output.clear_output()
clear_button = widgets.Button(description="Limpa")
clear_button.on_click(clear_button_handle)

#
# Desenho do "applet"
#
ui = widgets.VBox([x_texto, 
                   y_texto, 
                   ci_values,
                   upload,
                   widgets.HBox([show_button,  clear_button]),
                   output])

#
# Corre o applet ui=User Interface
#
ui
```
